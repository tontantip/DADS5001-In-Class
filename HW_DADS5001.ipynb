```python
def create_slide_5():
    team_logos_base64 = {}
    headers = {
        'Referer': 'https://www.premierleague.com/',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    for team, url in team_logos.items():
        if team not in squad_goalkeeping['Squad'].values:
            print(f"Skipping logo for '{team}' as it's not in the main data's 'Squad' list.")
            continue
        try:
            response = requests.get(url, headers=headers, timeout=10)
            if response.status_code == 200:
                encoded_image = base64.b64encode(response.content).decode('utf-8')
                data_uri = f"data:image/png;base64,{encoded_image}"
                team_logos_base64[team] = data_uri
            else:
                team_logos_base64[team] = None
        except Exception as e:
            team_logos_base64[team] = None
    print("Logo conversion complete.")


    # --- 4. ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Top 5 ---
    top_5_ga = squad_goalkeeping.sort_values(by='GA', ascending=True).head(5)
    top_5_cs = squad_goalkeeping.sort_values(by='CS', ascending=False).head(5)

    # --- 5. ‡∏™‡∏£‡πâ‡∏≤‡∏á Figure ---
    fig = make_subplots(
        rows=2, cols=1,
        subplot_titles=(
            "<b>Top 5 ‡∏ó‡∏µ‡∏°‡πÄ‡∏™‡∏µ‡∏¢‡∏õ‡∏£‡∏∞‡∏ï‡∏π‡∏ô‡πâ‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î (Total GA)</b>",
            "<b>Top 5 ‡∏ó‡∏µ‡∏°‡πÄ‡∏Å‡πá‡∏ö‡∏Ñ‡∏•‡∏µ‡∏ô‡∏ä‡∏µ‡∏ï‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î (Total CS)</b>"
        )
    )

    # --- 6. [‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç] ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Å‡∏£‡∏≤‡∏ü‡πÅ‡∏ó‡πà‡∏á (‡∏•‡∏ö text ‡∏≠‡∏≠‡∏Å) ---
    fig.add_trace(
        go.Bar(
            y=top_5_ga['Squad'],
            x=top_5_ga['GA'],
            # text=... (‡∏•‡∏ö‡∏≠‡∏≠‡∏Å)
            # textposition=... (‡∏•‡∏ö‡∏≠‡∏≠‡∏Å)
            orientation='h',
            name='Total GA',
            marker_color='#d62728',
            marker_opacity=0.8
        ),
        row=1, col=1
    )

    fig.add_trace(
        go.Bar(
            y=top_5_cs['Squad'],
            x=top_5_cs['CS'],
            # text=... (‡∏•‡∏ö‡∏≠‡∏≠‡∏Å)
            # textposition=... (‡∏•‡∏ö‡∏≠‡∏≠‡∏Å)
            orientation='h',
            name='Total CS',
            marker_color='#2ca02c',
            marker_opacity=0.8
        ),
        row=2, col=1
    )

    # --- 7. [‡πÉ‡∏´‡∏°‡πà] ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡∏Ç‡∏ô‡∏≤‡∏î‡∏†‡∏≤‡∏û‡πÅ‡∏•‡∏∞‡∏£‡∏∞‡∏¢‡∏∞‡∏´‡πà‡∏≤‡∏á ---
    image_width = 0.5  # ‡∏õ‡∏£‡∏±‡∏ö‡∏Ç‡∏ô‡∏≤‡∏î‡πÉ‡∏´‡πâ‡πÄ‡∏•‡πá‡∏Å‡∏•‡∏á‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÅ‡∏Å‡∏ô X ‡∏ó‡∏µ‡πà‡∏ô‡πâ‡∏≠‡∏¢
    text_offset = 0.1  # ‡∏õ‡∏£‡∏±‡∏ö‡∏£‡∏∞‡∏¢‡∏∞‡∏´‡πà‡∏≤‡∏á

    # --- 8. [‡πÉ‡∏´‡∏°‡πà] ‡∏ß‡∏ô Loop ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÇ‡∏•‡πÇ‡∏Å‡πâ‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (Subplot 1: GA) ---
    for i, row in top_5_ga.iterrows():
        team = row['Squad']
        value = row['GA']
        if team in team_logos_base64 and team_logos_base64[team]:
            fig.add_layout_image(
                dict(
                    source=team_logos_base64[team],
                    xref="x1", yref="y1",
                    x=value, y=team,
                    sizex=image_width, sizey=0.8,
                    xanchor="left", yanchor="middle"
                )
            )
        fig.add_annotation(
            xref="x1", yref="y1",
            x=value + image_width + text_offset,
            y=team,
            text=f"<b>{value}</b>",
            showarrow=False,
            xanchor="left", yanchor="middle",
            font=dict(color='black', size=12)
        )

    # --- 9. [‡πÉ‡∏´‡∏°‡πà] ‡∏ß‡∏ô Loop ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÇ‡∏•‡πÇ‡∏Å‡πâ‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (Subplot 2: CS) ---
    for i, row in top_5_cs.iterrows():
        team = row['Squad']
        value = row['CS']
        if team in team_logos_base64 and team_logos_base64[team]:
            fig.add_layout_image(
                dict(
                    source=team_logos_base64[team],
                    xref="x2", yref="y2",
                    x=value, y=team,
                    sizex=image_width, sizey=0.8,
                    xanchor="left", yanchor="middle"
                )
            )
        fig.add_annotation(
            xref="x2", yref="y2",
            x=value + image_width + text_offset,
            y=team,
            text=f"<b>{value}</b>",
            showarrow=False,
            xanchor="left", yanchor="middle",
            font=dict(color='black', size=12)
        )

    # --- 10. ‡∏õ‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏á Layout ---
    fig.update_layout(
        title_text="<b>Slide 5: ‡πÄ‡∏à‡∏≤‡∏∞‡∏•‡∏∂‡∏Å‡πÄ‡∏Å‡∏°‡∏£‡∏±‡∏ö (Defensive Strength)</b>",
        title_x=0.5,
        height=700,
        showlegend=False,
        yaxis=dict(autorange='reversed'),
        yaxis2=dict(autorange='reversed'),
        margin=dict(l=120, r=40, t=100, b=40),
        font=dict(family="Arial, sans-serif", size=12)
    )

    # --- 11. [‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç] ‡∏õ‡∏£‡∏±‡∏ö‡∏ä‡πà‡∏ß‡∏á‡∏Ç‡∏≠‡∏á‡πÅ‡∏Å‡∏ô X (‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì padding) ---
    max_ga = top_5_ga['GA'].max()
    max_cs = top_5_cs['CS'].max()
    padding = image_width + text_offset + 0.5 # ‡πÄ‡∏û‡∏¥‡πà‡∏° buffer

    fig.update_xaxes(range=[0, max_ga + padding], row=1, col=1)
    fig.update_xaxes(range=[0, max_cs + padding], row=2, col=1)

    # ‡πÅ‡∏™‡∏î‡∏á‡∏Å‡∏£‡∏≤‡∏ü
    return(fig.show())
```
