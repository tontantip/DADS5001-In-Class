```python
premier_league_table['xG_p90'] = premier_league_table['xG'] / premier_league_table['MP']
premier_league_table['xGA_p90'] = premier_league_table['xGA'] / premier_league_table['MP']

# 2. หาค่าเฉลี่ยของลีก
avg_xg_p90 = premier_league_table['xG_p90'].mean()
avg_xga_p90 = premier_league_table['xGA_p90'].mean()
fig = go.Figure()

# --- 1. เพิ่ม Scatter Trace ที่ "มองไม่เห็น" (สำหรับ Hover) ---
fig.add_trace(go.Scatter(
    x=premier_league_table['xG_p90'],
    y=premier_league_table['xGA_p90'],
    mode='markers',
    text=premier_league_table['Squad'],
    hovertemplate="<b>%{text}</b><br>xG (ต่อ 90): %{x:.2f}<br>xGA (ต่อ 90): %{y:.2f}<extra></extra>",
    marker=dict(size=10, opacity=0) # ทำให้จุดล่องหน
))

# --- 2. วน Loop เพื่อวาดโลโก้ ---
logo_size_x = 0.08 # ขนาดโลโก้ในแกน X
logo_size_y = 0.08 # ขนาดโลโก้ในแกน Y

for i, row in premier_league_table.iterrows():
    team = row['Squad']
    if team in team_logos_base64 and team_logos_base64[team]:
        fig.add_layout_image(
            dict(
                source=team_logos_base64[team],
                xref="x", yref="y",
                x=row['xG_p90'],
                y=row['xGA_p90'],
                sizex=logo_size_x,
                sizey=logo_size_y,
                xanchor="center", yanchor="middle"
            )
        )

# --- 3. เพิ่มเส้นค่าเฉลี่ย (เพื่อแบ่ง 4 ส่วน) ---
fig.add_vline(
    x=avg_xg_p90,
    line=dict(color='grey', width=2, dash='dot'),
    annotation_text='ค่าเฉลี่ยเกมรุก'
)
fig.add_hline(
    y=avg_xga_p90,
    line=dict(color='grey', width=2, dash='dot'),
    annotation_text='ค่าเฉลี่ยเกมรับ'
)

# --- 4. ปรับแต่ง Layout ---
fig.update_layout(
    title_text="<b>วิเคราะห์ความครบเครื่อง (เกมรุก vs เกมรับ)</b>",
    title_x=0.5,
    xaxis_title="เกมรุก (xG per 90)  → (ยิ่งขวายิ่งดี)",
    yaxis_title="เกมรับ (xGA per 90)  → (ยิ่งล่างยิ่งดี)",
    height=700,
    width=800,
    showlegend=False,
    hovermode="closest",

    # [สำคัญ] พลิกแกน Y (Invert Y-Axis) เพื่อให้ค่าน้อย (รับดี) อยู่ด้านบน
    # *** Plotly ทำให้การกลับแกนซับซ้อนเล็กน้อย วิธีที่ง่ายกว่าคือ
    #     อ่านค่า "ล่างขวา" ว่าดีที่สุด แทนที่จะกลับแกนครับ

    # (ถ้าอยากกลับแกน Y จริงๆ)
    # yaxis=dict(autorange="reversed"),

    xaxis=dict(zeroline=False),
    yaxis=dict(zeroline=False),

    margin=dict(l=40, r=40, t=100, b=40)
)

# เพิ่มคำอธิบาย 4 ส่วน
fig.add_annotation(
    x=premier_league_table['xG_p90'].max(), y=avg_xga_p90,
    xref='x', yref='y domain', yanchor='bottom', xanchor='right',
    text="🏆 <b>ครบเครื่อง (Elite)</b>", font=dict(color='green'), showarrow=False
)
fig.add_annotation(
    x=premier_league_table['xG_p90'].min(), y=avg_xga_p90,
    xref='x', yref='y domain', yanchor='bottom', xanchor='left',
    text="🛡️ <b>เน้นรับ (Defensive)</b>", font=dict(color='blue'), showarrow=False
)
fig.add_annotation(
    x=premier_league_table['xG_p90'].max(), y=avg_xga_p90,
    xref='x', yref='y domain', yanchor='top', xanchor='right',
    text="⚔️ <b>เน้นบุก (Attacking)</b>", font=dict(color='orange'), showarrow=False
)
fig.add_annotation(
    x=premier_league_table['xG_p90'].min(), y=avg_xga_p90,
    xref='x', yref='y domain', yanchor='top', xanchor='left',
    text="📉 <b>มีปัญหา (Struggling)</b>", font=dict(color='red'), showarrow=False
)

fig.show()
```
